```python
import sympy as sm
from sympy import symbols, Matrix, solve, simplify
import sympy.physics.mechanics as me
me.init_vprinting()
q1, q2, q3, u1, u2, u3 = me.dynamicsymbols('q1:4, u1:4')
l1, l2, l3, l4, rho = sm.symbols('l1:5, rho')
m1, m2, m3, m4 = sm.symbols('m1, m2, m3, m4')
```

```python
N = me.ReferenceFrame('N')
inertias = [me.inertia(N, 0, 0, m) for l, m in zip((l1, l2, l3, l4), (m1, m2, m3, m4))]
link1 = me.Body('A', frame=N, mass=m1, central_inertia=inertias[0])
link2 = me.Body('B', mass=m2, central_inertia=inertias[1])
link3 = me.Body('C', mass=m3, central_inertia=inertias[2])
link4 = me.Body('D', mass=m4, central_inertia=inertias[3])
```

```python
joint1 = me.PinJoint('J1', link1, link2, coordinates=q1, speeds=u1,
                  parent_point=l1 / 2 * link1.x,
                  child_point=-l2 / 2 * link2.x, joint_axis=link1.z)
joint2 = me.PinJoint('J2', link2, link3, coordinates=q2, speeds=u2,
                  parent_point=l2 / 2 * link2.x,
                  child_point=-l3 / 2 * link3.x, joint_axis=link2.z)
joint3 = me.PinJoint('J3', link3, link4, coordinates=q3, speeds=u3,
                  parent_point=l3 / 2 * link3.x,
                  child_point=-l4 / 2 * link4.x, joint_axis=link3.z)
```

```python
loop = link4.masscenter.pos_from(link1.masscenter) + l1 / 2 * link1.x + l4 / 2 * link4.x
fh = Matrix([loop.dot(link1.x).simplify(), loop.dot(link1.y).simplify()])
fh
```

```python
T = me.dynamicsymbols("T")
link2.apply_torque(T*link1.frame.z)
link2.loads
```

```python
method = me.JointsMethod(link1, joint1, joint2, joint3)
t = me.dynamicsymbols._t
qdots = solve(method.kdes, [q1.diff(t), q2.diff(t), q3.diff(t)])
fhd = fh.diff(t).subs(qdots)
fhd
```

```python
fhdd = fhd.diff(t).subs(qdots)
fhdd
```

```python
method.loads
```

```python
method._method = me.KanesMethod(
    method.frame, 
    q_ind=[q1], 
    u_ind=[u1], 
    q_dependent=[q2, q3],
    u_dependent=[u2, u3], 
    kd_eqs=method.kdes,
    configuration_constraints=fh, 
    velocity_constraints=fhd,
    acceleration_constraints=fhdd, 
    bodies=method.bodies)

# sim = simplify(method.method._form_eoms())
```

```python
fr, frstar = method.method.kanes_equations(loads=method.loads)
zero = fr + frstar
zero
```

```python
method.loads
```

```python
me.find_dynamicsymbols(zero)
```

```python
me.find_dynamicsymbols(method.method.forcing_full)
```

```python
import numpy as np
from scipy.optimize import fsolve
from pydy.system import System
```

```python
sys = System(method.method)
```

```python
sys.constants_symbols
```

```python
sys.specifieds_symbols
```

```python
sys.initial_conditions = {q1: 10,
                          q2: -6.57,
                          q3: 151,
                          u1: 0.0,
                          u2: 0.0,
                          u3: 0.0}
```

```python
def step_pulse(x,y):
    return np.array([1.0])
sys.specifieds = {T: step_pulse}
```

```python
l1_val = 0.8
l2_val = 2.0
l3_val = 1.0
l4_val = 2.0

sys.constants = {l2: l2_val,
                 l3: l3_val,
                 l4: l4_val,
                 m2: 1.0,
                 m3: 1.0,
                 m4: 1.0}
```

```python
duration = 4.0
fps = 60.0
sys.times = np.linspace(0.0, duration, num=int(fps*duration))
x = sys.integrate()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(3, 2, sharex=True)
fig.set_size_inches(10, 10)

for i, (xi, ax, s) in enumerate(zip(x.T, axes.T.flatten(), sys.states)):
    ax.plot(sys.times, np.rad2deg(xi))
    title = sm.latex(s, mode='inline')
    ax.set_title(title)
    if 'q' in title:
        ax.set_ylabel('Angle [deg]')
    else:
        ax.set_ylabel('Angular Rate [deg/s]')

axes[2, 0].set_xlabel('Time [s]')
axes[2, 1].set_xlabel('Time [s]')

plt.tight_layout()
```
